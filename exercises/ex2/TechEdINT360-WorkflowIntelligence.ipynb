## SAP Workflow Management: Introduction
SAP Workflow Management allows you to digitize workflows, manage decisions, gain end-to-end process visibility, and configure processes in a low-code approach. It allows users to build, run, and manage workflows. It enables users to digitize and automate decision making. It enables process excellence, process transparency, process transformation by providing one view of the process. Users can configure the process flow without the involvement of IT to improve process efficiency. Please find the documentation [link here](https://help.sap.com/viewer/product/WORKFLOW_MANAGEMENT/Cloud/en-US?task=discover_task)

### Workflow Intelligence 
Workflow Intelligence is a capability of SAP Workflow Management that allows Business Users to configure Machine Learning based Workflow Intelligence scenarios to recommend, predict and identify bottlenecks in a business process.

#### Step 2.2.2.1
Import the SalesOrderDataForTechEd2021.csv file from the data folder and create a dataframe

```python
# if you are experiencing an import erro, run the previous cell to install panda by uncommenting the install command
import pandas as pd

df_sales_orders = pd.read_csv("data/SalesOrderDataForTechEd2021.csv")
df_sales_orders.head(5)
```

#### Step 2.2.2.2 
Extract information required to generate an access token

```python
#getting the url from service key you have imported earlier which will be used for authentication
uaa_url = SERVICE_KEY.get('uaa').get('url')+'/oauth/token'
client_id = SERVICE_KEY.get('uaa').get('clientid')
client_secret =SERVICE_KEY.get('uaa').get('clientsecret')

print("Token URL :"+ uaa_url)
#print("Client Id :"+ client_id)
#print("Client secret :"+client_secret)
```

#### Step 2.2.2.3
Request for authorization token

```python
#get oauth token and create authorization header for request
import requests

grant_type = "client_credentials"
data = {
    "grant_type": grant_type,
    "client_id": client_id,
    "client_secret": client_secret,
    }
auth_response = requests.post(uaa_url, data=data)
auth_response
# Read token from auth response
auth_response_json = auth_response.json()
auth_token = auth_response_json["access_token"]
auth_token_header_value = "Bearer %s" % auth_token
auth_token_header = {"Authorization": auth_token_header_value}

```

```python
#uncomment if you want to copy the token manually
#print ("Token :"+auth_token)
```

#### Step 2.2.2.4
In this step we will instantiate some workflows and complete them. By doing this we will get a dataset of completed workflows with historical decision.

Extract the base URL for Workflow API calls

```python
#get the REST API url for calling the workflow apis
workflow_rest_api_url=SERVICE_KEY.get('endpoints').get('workflow_rest_url')

print("Workflow rest api url : " + str(workflow_rest_api_url))
```

#### Step 2.2.2.5
Define the workflow and task definition which will be used to configure the Workflow Intelligence scenario

```python
#store the workflow definition id and task definition id
workflow_definition = "com.sap.bpm.wfi.approvesalesorder_teched21_int360"
task_definition = "firstApprovalForSO"
```

#### Step2.2.2.6
Prepare the URL parts for starting set of workflows for approval and rejection 


```python
start_query_param = "/v1/workflow-instances"
start_workflow_url = workflow_rest_api_url+start_query_param
headers={
    'Content-type':'application/json', 
    'Authorization': auth_token_header_value
}
```

#### Step2.2.2.7
Start some workflows to approve. We are creating a set of workflow that will be approved and completed to use them for training purpose. In a productive scenario we rely on already completed workflows.



```python
#start sales order approval workflows to be Approved
#first we start all the workflows which are to be approved in the dataset

df_approve_so=df_sales_orders[df_sales_orders['decision']=='Approved']

for i in range(len(df_approve_so)):

    temp_df = df_approve_so.iloc[i].filter(['wfi_SalesOrder','wfi_SalesOrganization','wfi_DistributionChannel','wfi_OrganizationDivision','wfi_SoldToParty','wfi_Product_1','Rate','wfi_P1_Req_Qty','wfi_SalesOrderDate','wfi_NetValue','wfi_P1_Req_Del_Date']).to_json()
    start_context = '{"definitionId":"'+workflow_definition+'","context":'+str(temp_df)+'}'
    #print ("Start Context for "+workflow_definition+" is \n" + start_context)
    
    #start workflow
    response = requests.post(start_workflow_url, headers=headers,data = start_context)
    
    # Consider any status other than 2xx an error
    if not response.status_code // 100 == 2:
        print ('Error: Unexpected response %s for definition %s' %(response.status_code,workflow_definition))
    else:
        print ('Started workflow to approve : %s'%(i))
```

#### Step2.2.2.8
Get all the tasks instance in the READY state

```python
#get all the tasks which are in READY state for APPROVAL 
url_suffix_get_ready_tasks = "/v1/task-instances"
url_get_ready_tasks = workflow_rest_api_url + url_suffix_get_ready_tasks +"?activityId="+task_definition+"&status=READY"
print("Query URL for Ready tasks:\n"+ url_get_ready_tasks)
response = requests.get(url_get_ready_tasks, headers=headers).json()
print(response)
```

#### Step2.2.2.9
Approve all the tasks which we queried in previous step

```python
#Approve all the tasks
url_complete_task_param1 = '/v1/task-instances/'
data_to_complete_task = '{"context":{"decision":"Approved"},"status":"COMPLETED"}'
for i in range(0,len(response)):
    temp_url = workflow_rest_api_url+url_complete_task_param1+response[i].get("id")
    #print(temp_url)
    response_task_completed = requests.patch(temp_url,headers=headers,data=data_to_complete_task)
    print(response[i].get("id")+" Task Approved - Response:"+str(response_task_completed)+"\n")
```

#### Step2.2.2.10
Start some workflows to reject to get a realistic data set.Now we are instantiating a set of workflow which will be rejected and completed so that we can use it for training purposes. In a productive scenario we rely on already completed workflows.

```python
#start sales order approval workflows to be Rejected
#we will start workflows to be Rejected in Dataset

df_reject_so=df_sales_orders[df_sales_orders['decision']=='Rejected']
#print(len(df_reject_so))

for i in range(len(df_reject_so)):
#for i in range(1,len(df_reject_so)):
    temp_df = df_reject_so.iloc[i].filter(['wfi_SalesOrder','wfi_SalesOrganization','wfi_DistributionChannel','wfi_OrganizationDivision','wfi_SoldToParty','wfi_Product_1','Rate','wfi_P1_Req_Qty','wfi_SalesOrderDate','wfi_NetValue','wfi_P1_Req_Del_Date']).to_json()
    start_context = '{"definitionId":"'+workflow_definition+'","context":'+str(temp_df)+'}'
    #print ("Start Context for "+workflow_definition+" is \n" + start_context)
    
    #start workflow
    response = requests.post(start_workflow_url, headers=headers,data = start_context)
    
    # Consider any status other than 2xx an error
    if not response.status_code // 100 == 2:
        print ('Error: Unexpected response %s for definition %s' %(response.status_code,workflow_definition))
    else:
        print ('Started workflow to reject : %s'%(i))
```

#### Step2.2.2.11
Get all the tasks instance in the READY state 

```python
#get all the tasks which are in READY state for REJECTION 
url_suffix_get_ready_tasks = "/v1/task-instances"
url_get_ready_tasks = workflow_rest_api_url + url_suffix_get_ready_tasks +"?activityId="+task_definition+"&status=READY"
print("Qurery URL for Ready tasks:\n"+ url_get_ready_tasks)

response = requests.get(url_get_ready_tasks, headers=headers).json()
print(response)
```

#### Step2.2.2.12
Reject all the tasks which we queried in previous step

```python
#Reject all the tasks
url_complete_task_param1 = '/v1/task-instances/'
data_to_complete_task = '{"context":{"decision":"Rejected"},"status":"COMPLETED"}'
for i in range(0,len(response)):
    temp_url = workflow_rest_api_url+url_complete_task_param1+response[i].get("id")
    #print(temp_url)
    response_task_completed = requests.patch(temp_url,headers=headers,data=data_to_complete_task)
    print(response[i].get("id")+" Task Rejected - Response:"+str(response_task_completed)+"\n")
```
